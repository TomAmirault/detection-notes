```python
import sounddevice as sd
import soundfile as sf
import whisper
import time
import os


def transcribe_audio(file_path, model_size="tiny"):
    model = whisper.load_model(model_size)
    result = model.transcribe(file_path)
    return result["text"]


def record_and_transcribe_loop(duration=10, samplerate=16000, model_size="tiny"):
    model = whisper.load_model(model_size)  

    try:
        counter = 1
        while True:
            filename = f"record_chunk_{counter}.wav"
            if counter == 1 :
                print(f"Parlez maintenant")
            recording = sd.rec(int(duration * samplerate), samplerate=samplerate, channels=1, dtype="float32")
            sd.wait()

            sf.write(filename, recording, samplerate)

            result = model.transcribe(filename, initial_prompt="1")
            print(f"Transcription [{counter}] :", result["text"], "\n")

            counter += 1

    except KeyboardInterrupt:
        print("\n Arrêt demandé par l'utilisateur (Ctrl+C).")


if __name__ == "__main__":
    record_and_transcribe_loop(duration=10, model_size="tiny")
```
